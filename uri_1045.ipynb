# ***Tipos de Triângulos***
Leia 3 valores de ponto flutuante A, B e C e ordene-os em ordem decrescente, de modo que o lado A representa o maior dos 3 lados. A seguir, determine o tipo de triângulo que estes três lados formam, com base nos seguintes casos, sempre escrevendo uma mensagem adequada:<br/><br/>
- se A ≥ B+C, apresente a mensagem: NAO FORMA TRIANGULO<br/>
- se A<sup>2</sup> = B<sup>2</sup> + C<sup>2</sup>, apresente a mensagem: TRIANGULO RETANGULO<br/>
- se A<sup>2</sup> > B<sup>2</sup> + C<sup>2</sup>, apresente a mensagem: TRIANGULO OBTUSANGULO<br/>
- se A<sup>2</sup> < B<sup>2</sup> + C<sup>2</sup>, apresente a mensagem: TRIANGULO ACUTANGULO<br/>
- se os três lados forem iguais, apresente a mensagem: TRIANGULO EQUILATERO<br/>
- se apenas dois dos lados forem iguais, apresente a mensagem: TRIANGULO ISOSCELES

<br/>***Entrada***<br/>
A entrada contem três valores de ponto flutuante de dupla precisão A (0 < A) , B (0 < B) e C (0 < C).

<br/>***Saída***<br/>
Imprima todas as classificações do triângulo especificado na entrada.

```
lados = [float(input()),float(input()),float(input())]
lados= sorted(lados)[::-1]

a = lados[0]
b = lados[1]
c = lados[2]

if a >= (b+c):
  print('NAO FORMA TRIANGULO')
if (a*a) == (b*b+c*c):
  print('TRIANGULO RETANGULO')
if (a*a) > (b*b+c*c):
  print('TRIANGULO OBTUSANGULO')
if (a*a) < (b*b+c*c):
  print('TRIANGULO ACUTANGULO') 
if (a == b and b == c):
  print('TRIANGULO EQUILATERO')
if (a == b and b!=c) or (a == c and b!=c) or (b == c and a!=c):
  print('TRIANGULO ISOSCELES')     
```
